```python
import my_linear_algebra as mla
data = mla.deepcopy(raw)
```

```python
# X 생성
X = []
for i in range(1,n+1):
    row = []
    for j in range(1,p):
        row.append(data[i][j])
    X.append(row)
```

```python
# y 생성
y = []
for i in range(1, n+1):
    val = int(data[i][0])
    y.append(val)
```

```python
# X 스케일 조정
X1 = mla.deepcopy(X)
n1 = len(X)
p1 = len(X[0])

X_scale = []
for i in range(0, n1):
    row = []
    for j in range(0, p1):
        val = float(X1[i][j])/255
        row.append(val)
    X_scale.append(row)
```

```python
# (10000, 28, 28, 1) 차원으로 변경
```

```python
print(X_scale[0])
```

```python
n = 10000
w = 28
h = 28
c = 1
```

```python
X_rs = []
for i in range(0, n):
    image = []
    tmp_image = mla.deepcopy(X_scale[i])
    for j in range(0, w):
        row = []
        for k in range(0, h):
            pixel = []
            for l in range(0, c):
                val = tmp_image.pop(l) 
                pixel.append(val)
            row.append(pixel)
        image.append(row)
    X_rs.append(image)
```

```python
y_set = set(y)
print(y_set)
```

```python
n_category = len(y_set)
print(n_category)
```

```python
y_onehot = []
for i in range(0, n):
    row = []
    for j in range(0, n_category):
        if y[i] == j:
            row.append(1)
        else:
            row.append(0)
    y_onehot.append(row)
```

```python
y_onehot
```

```python
X = X_rs
y = y_onehot
```

```python
# 트레이닝/테스트 데이터 분할
from sklearn.model_selection import train_test_split
X_tn, X_te, y_tn, y_te=train_test_split(X,y,random_state=0)
```

```python
# 합성곱 신경망 생성
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D
from tensorflow.keras.layers import MaxPool2D, Flatten
from tensorflow.keras.layers import Dropout

n_class = len(y_tn[0])

model = Sequential()
model.add(Conv2D(32, kernel_size=(5,5), 
                 input_shape=(28,28,1), 
                 padding='valid', 
                 activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Conv2D(32, kernel_size=(3,3), 
                 padding='valid', 
                 activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(1024, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(n_class, activation='softmax'))
model.summary()
```

```python
# 모형 컴파일
model.compile(loss='categorical_crossentropy', 
              optimizer='adam', 
              metrics=['accuracy'])
```

```python
hist = model.fit(X_tn, y_tn, epochs=3, batch_size=100)
```

```python
print(model.evaluate(X_tn, y_tn)[1])
```

```python
# 테스트 데이터 평가
print(model.evaluate(X_te, y_te)[1])
```
